```pyspark
from graphframes import *
from pyspark.sql import *
from pyspark.sql import SparkSession
```

```pyspark
spark = SparkSession.builder.appName('abc').getOrCreate()
input_bucket = 's3://bucket-for-graph'
input_path1 = '/e-30RU9I5GXV5US42PNFOP4WOY3/item.csv'
input_path2 = '/e-30RU9I5GXV5US42PNFOP4WOY3/statements.csv'
item_df = spark.read.csv(input_bucket + input_path1, header=True)
item_df.show()
#statementdf=spark.read.csv('statements.csv',header=True)
#s3://bucket-for-graph/e-30RU9I5GXV5US42PNFOP4WOY3/item.csv
#itemdf=spark.read.csv('item.csv',header=True)
```

```pyspark
statement_df = spark.read.csv(input_bucket + input_path2, header=True)
statement_df.show()
```

```pyspark
columns_to_drop = ['en_description']
vertices = item_df.drop(*columns_to_drop)
vertices = vertices.selectExpr("item_id as id", "en_label as name")
edges= statement_df
edges = edges.selectExpr("source_item_id as src","target_item_id as dst")

#g = GraphFrame(vertices, edges)
```

```pyspark
g = GraphFrame(vertices, edges)
```

```pyspark
def get_length(paths):
    length = 0
    for i in paths:
        length += 1
    length = (length - 1)/2
    return int(length)

def find_shortest_path_length(i,j):
    paths = g.bfs("id = "+str(i),"id = " + str(j))
    return get_length(paths)
```

```pyspark
a = find_shortest_path_length(1,497745)
a
```

```pyspark
%%time
b = find_shortest_path_length(1,152)
b
```
